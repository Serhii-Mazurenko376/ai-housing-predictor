```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Exempeldata: [kvadratmeter, antal rum]
X = np.array([
    [30, 1],
    [45, 2],
    [60, 3],
    [80, 4],
    [100, 5]
])

# Priser i tusentals kronor
y = np.array([1200, 1800, 2500, 3200, 4000])

# Skapa och träna modellen
model = LinearRegression()
model.fit(X, y)

# Test: Gör en förutsägelse för 70 kvm, 3 rum
new_apartment = np.array([[70, 3]])
predicted_price = model.predict(new_apartment)

print(f"Förväntat pris för 70 kvm, 3 rum: {int(predicted_price[0])} tusen kr")
```

```python
import matplotlib.pyplot as plt

# Förutsägelser för alla datapunkter
y_pred = model.predict(X)

# Rita scatterplot och linje
plt.scatter(X[:, 0], y, color='blue', label='Faktiska priser')
plt.plot(X[:, 0], y_pred, color='red', label='Modellens förutsägelser')
plt.xlabel('Kvadratmeter')
plt.ylabel('Pris (tusen kr)')
plt.title('Bostadspriser vs. Storlek')
plt.legend()
plt.grid(True)
plt.show()
```
